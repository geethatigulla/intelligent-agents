```python
# Verify imports
import chromadb
from sentence_transformers import SentenceTransformer
from litellm import completion

print("✅ All imports successful!")
```

```python
# Load a pre-trained embedding model
print("Loading embedding model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print(f"✅ Model loaded! Dimension: {model.get_sentence_embedding_dimension()}")
```

```python
# Exercise 2.1: Create embeddings

texts = [
    "I love drinking coffee in the morning",
    "I enjoy having espresso after lunch",
    "The stock market crashed yesterday",
    "Python is a great programming language"
]

# Use model.encode() to create embeddings
embeddings = model.encode(texts)

print(f"Created {len(embeddings)} embeddings")
print(f"Each embedding has {len(embeddings[0])} dimensions")
```

```python
# Calculate similarity between texts
from sentence_transformers import util

print("📊 Similarity between texts:\n")

# Compare coffee texts
sim_coffee = util.cos_sim(embeddings[0], embeddings[1]).item()
print(f"'coffee' vs 'espresso': {sim_coffee:.3f}")

# Compare coffee vs stock market
sim_diff = util.cos_sim(embeddings[0], embeddings[2]).item()
print(f"'coffee' vs 'stock market': {sim_diff:.3f}")

print("\n💡 Higher number = more similar!")
```

```python
# Exercise 2.2: Semantic Search

query = "What's a good hot beverage?"

# Step 1: Create embedding for the query
query_embedding = model.encode(query)

# Step 2: Calculate similarities
similarities = util.cos_sim(query_embedding, embeddings)[0]

# Step 3: Find the best match
best_idx = similarities.argmax().item()
best_text = texts[best_idx]
best_score = similarities[best_idx].item()

print(f"🔍 Query: '{query}'")
print(f"🥇 Best match: '{best_text}'")
print(f"   Similarity: {best_score:.3f}")
```

```python
# Exercise 4.1: Build a RAG function

def simple_rag(question: str, collection, n_results: int = 2) -> str:
    """
    Simple RAG implementation:
    1. Retrieve relevant documents
    2. Build context from documents
    3. Generate answer using LLM
    """

    # Step 1: Retrieve relevant documents
    results = collection.query(
        query_texts=[question],
        n_results=n_results
    )

    # Step 2: Build context from retrieved documents
    context = "\n".join(results['documents'][0])

    # Step 3: Create the prompt with context
    prompt = f"""Based on the following context, answer the question.

CONTEXT:
{context}

QUESTION: {question}

ANSWER:"""

    # Step 4: Generate answer with LLM
    response = completion(
        model="groq/llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=300
    )

    return response.choices[0].message.content

print("✅ RAG function created!")
```

```python
# Test the RAG function
question = "What is machine learning?"

print(f"❓ Question: {question}\n")
print("⏳ Processing...\n")

answer = simple_rag(question, collection)

print(f"🤖 Answer:\n{answer}")
```

```python
# Create sample documentation
documentation = [
    "SuperApp Installation: Download from our website. Requires Windows 10+, macOS 12+, or Ubuntu 20.04+.",
    "SuperApp Pricing: Free tier has 100 API calls/day. Premium is $29/month with unlimited calls.",
    "SuperApp Login Issues: Check internet connection. Try 'Forgot Password'. Clear browser cache.",
    "SuperApp API: Use REST API with API keys from Settings. Rate limit is 100 requests/minute.",
    "SuperApp Export: Export data in CSV, JSON, or PDF format via Data > Export menu."
]

print(f"📚 Created {len(documentation)} documentation entries")
```

```python
# Create a new collection for documentation
try:
    client.delete_collection("superapp_docs")
except:
    pass

docs_collection = client.create_collection(name="superapp_docs")

docs_collection.add(
    documents=documentation,
    ids=[f"doc_{i}" for i in range(len(documentation))]
)

print(f"✅ Added {docs_collection.count()} documents to knowledge base")
```

```python
# Exercise 5.1: Build the Q&A Agent class

class DocumentationQAAgent:
    """A Q&A agent that answers questions using RAG."""

    def __init__(self, collection):
        self.collection = collection
        self.system_prompt = """You are a helpful documentation assistant.
Answer questions based on the provided documentation.
Be concise and helpful. If unsure, say so."""

    def answer(self, question: str) -> str:
        """Answer a question using RAG."""

        # Step 1: Retrieve relevant docs
        results = self.collection.query(
            query_texts=[question],
            n_results=2
        )

        # Step 2: Build context
        context = "\n".join(results['documents'][0])

        # Step 3: Create messages for LLM
        messages = [
            {"role": "system", "content": self.system_prompt},
            {"role": "user", "content": f"DOCS:\n{context}\n\nQUESTION: {question}"}
        ]

        # Step 4: Generate response
        response = completion(
            model="groq/llama-3.1-8b-instant",
            messages=messages,
            max_tokens=300
        )

        return response.choices[0].message.content

# Create the agent
qa_agent = DocumentationQAAgent(docs_collection)
print("✅ Q&A Agent created!")
```

```python
# Test the Q&A Agent
questions = [
    "How do I install SuperApp?",
    "How much does premium cost?",
    "I can't log in, what should I do?"
]

for q in questions:
    print(f"\n{'='*50}")
    print(f"❓ User: {q}\n")

    answer = qa_agent.answer(q)
    print(f"🤖 Agent: {answer}")
```

```python
# Exercise 6.1: Text chunking

def chunk_text(text: str, chunk_size: int = 100) -> list:
    """
    Split text into chunks of approximately chunk_size characters.
    """
    chunks = []

    # Split into sentences (simple approach)
    sentences = text.replace('!', '.').replace('?', '.').split('.')

    current_chunk = ""

    for sentence in sentences:
        sentence = sentence.strip()
        if not sentence:
            continue

        # Check if adding this sentence exceeds chunk_size
        if len(current_chunk) + len(sentence) > chunk_size:
            if current_chunk:
                chunks.append(current_chunk.strip())
            current_chunk = sentence + "."
        else:
            current_chunk += " " + sentence + "."

    # Don't forget the last chunk
    if current_chunk:
        chunks.append(current_chunk.strip())

    return chunks
```

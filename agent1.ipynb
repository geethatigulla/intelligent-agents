```python
# Create calculator tool - fill in the blanks
                                    # What decorator?
def calculator(expression: str) -> str:
    """Evaluates math expressions. Use for calculations.

    Args:
        expression: Math like '2+2' or '100*5'
    """
    try:
        result = eval(expression)
        return str(result)
    except Exception as e:
        return f"Error: {e}"

# âœ… Checkpoint
print("âœ… Calculator tool created!")
```

```python
# Test calculator DIRECTLY - no agent, no API needed!
print("ğŸ§® Testing calculator directly...\n")

result1 = calculator("12345 * 67890")
print(f"12345 Ã— 67890 = {result1}")

result2 = calculator("(100 + 50) * 2")
print(f"(100 + 50) Ã— 2 = {result2}")

print("\nâœ… Tool works perfectly!")
```

```python
from litellm import completion

class ConversationAgent:
    """Agent that remembers conversation."""

    def __init__(self):
        self.history = []  # Empty list to store messages
        print("ğŸ¤– Agent with memory created!")

    def chat(self, message):
        # Add user message to history
        self.history.append({"role": "user", "content": message})

        # Send FULL history to LLM
        response = completion(
            model="groq/llama-3.1-8b-instant",
            messages=self.history  # What contains all messages?
        )

        # Extract response text
        assistant_message = response.choices[0].message.content

        # Save response to history
        self.history.append({"role": "assistant", "content": assistant_message})
        return assistant_message

    def show_memory(self):
        print("\nğŸ“ Memory:")
        for i, msg in enumerate(self.history, 1):
            role = "You" if msg["role"] == "user" else "Agent"
            print(f"{i}. [{role}]: {msg['content'][:50]}...")

print("âœ… Class created!")
```

```python
# Create instance
memory_agent = ConversationAgent()
```

```python
# Message 1
print("--- Message 1 ---")
r1 = memory_agent.chat("My name is Ravi and I am a software engineer.")
print("Agent:", r1)
```

```python
# Message 2 - Does it remember?
print("--- Message 2 ---")
r2 = memory_agent.chat("What is my name and profession?")
print("Agent:", r2)
print("\nğŸ‰ It remembers!")
```

```python
from smolagents import ToolCallingAgent, LiteLLMModel, DuckDuckGoSearchTool
from litellm import completion

class ITHelpdeskAgent:
    def __init__(self, user_name=None, enable_search=False):
        self.user_name = user_name
        self.enable_search = enable_search
        self.history = []
        self.system_prompt = f"""
You are TechBot, friendly IT helpdesk assistant.
STYLE: Numbered steps, end with "- TechBot ğŸ¤–"
RULES: No passwords. Hardware â†’ Room 105.
{f"User: {user_name}" if user_name else ""}
"""
        # Only create model/tools if search is enabled
        if enable_search:
            self.model = LiteLLMModel(model_id="groq/llama-3.1-8b-instant")
            self.tools = [DuckDuckGoSearchTool()]

        print(f"ğŸ¤– TechBot ready!")
        if user_name: print(f"   Hi {user_name}!")
        if enable_search: print("   ğŸ” Search enabled (may hit rate limits)")

    def chat(self, message):
        self.history.append({"role": "user", "content": message})

        if self.enable_search:
            # Use ToolCallingAgent for search (OPTIONAL - may hit rate limits)
            if len(self.history) > 1:
                prompt = "CONVERSATION:\n" + "\n".join(
                    f"{'USER' if m['role']=='user' else 'TECHBOT'}: {m['content']}"
                    for m in self.history
                )
            else:
                prompt = message
            agent = ToolCallingAgent(tools=self.tools, model=self.model,
                              system_prompt=self.system_prompt, verbosity_level=0, max_steps=3)
            response = agent.run(prompt)
        else:
            # Use direct litellm.completion() - more reliable!
            messages = [{"role": "system", "content": self.system_prompt}]
            messages.extend(self.history)
            result = completion(model="groq/llama-3.1-8b-instant", messages=messages)
            response = result.choices[0].message.content

        self.history.append({"role": "assistant", "content": response})
        return response

    def show_history(self):
        print("\nğŸ“œ History:")
        for m in self.history:
            icon = "ğŸ‘¤" if m["role"]=="user" else "ğŸ¤–"
            print(f"{icon} {m['content'][:70]}...\n")

print("âœ… ITHelpdeskAgent ready!")
```
